<h1> Créer sa première figure </H1>

<b> Définition d'une figure </b>

La "Figure" est le conteneur de niveau supérieur dans une visualisation Matplotlib. Elle inclut tout l'espace de la visualisation, y compris les axes, graphiques, textes, labels, etc. Pensez à la figure comme à la "toile" sur laquelle tout est dessiné.

```python
# Exemple de code pour créer une figure vide
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure() #creer la figure

plt.show() # afficher la courbe 
```

<b> Ajout d'axes à la figure </b>

Les axes représentent la zone de dessin de votre graphique

```python
fig = plt.figure()
ax = fig.add_subplot(111)  # 111 signifie une grille de 1x1 et nous sélectionnons le premier subplot.
plt.show()
```

<b> Tracé de Données Basiques </b>

```python
x = np.linspace(0, 10, 100)  # 100 points de 0 à 10
y = np.sin(x) # y est le sinus de x
print(x)
print(y)
```

```python
#Utilisez les données pour tracer une courbe (plot)
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(x, y)
plt.show()
```

<b> Personnalisation du Graphique </b>

```python
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(x, y)

ax.set_title("Graphique du sinus") # un titre donne un contexe au graphique
ax.set_xlabel("x")  # Il est essentiel d'etiqueter vos axes pour indiquer ce que les donnees representent
ax.set_ylabel("sin(x)") 

plt.show()
```

<h1> Création de plusieurs graphiques sur une figure </h1>

<b> Créer 2 courbes sur un même graphique </b>

```python
x = np.linspace(0, 10, 100)  # 100 points de 0 à 10
y1 = np.sin(x)                # y est le sinus de x
y2 = np.cos(x)                # y est le cosinus de x
```

```python
# Affichage de 2 courbes sur le même grahique
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x , y1 , label = 'sinus')
ax.plot(x , y2 , label = 'cosinus')
ax.set_title('Graphique du sinus et cosinus')
ax.legend()
plt.show()
```

<b> Afficher les courbes sur 2 graphiques différents </b>

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.set_title("Sinus")
ax1.plot(x,y1, color='lightblue')

ax2= fig.add_subplot(122)
ax2.set_title("Cosinus")
ax2.plot(x,y2, color='darkgreen')
plt.show()
```

```python
fig = plt.figure(figsize=(7,7)) # Ajuster la taille pour éviter les supperpositions
ax1 = fig.add_subplot(221) # 2 lignes, 2 colonnes, position 1
ax1.set_title("en haut à gauche") 
ax1.plot(x,y1, color='lightblue')

ax2 = fig.add_subplot(222) # 2 lignes, 2 colonnes, position 2
ax2.set_title("en haut à droite") 
ax2.plot(x,y2, color='darkgreen')

ax3 = fig.add_subplot(223) # 2 lignes, 2 colonnes, position 3
ax3.set_title("en bas à gauche") 
ax3.scatter(x,y1, color='lightblue')

ax4 = fig.add_subplot(224) # 2 lignes, 2 colonnes, position 4
ax4.set_title("en bas à droite") 
ax4.scatter(x,y2, color='darkgreen') # scater c'est un nuage de points

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(x, y)
ax.set_title("Graphique du Sinus")
ax.set_xlabel("x") 
ax.set_ylabel("sin(x)")

plt.savefig("mes_graphiques.png") # Sauvegarde format png du graphique
plt.show()
```

<h1> Anatomie d'une figure </h1>



- Figure: La figure représente l'ensemble de l'espace de visualisation. C'est le conteneur principal dans lequel tous les autres éléments graphiques sont ajoutés, comme les axes, les graphiques, les textes, et les légendes.
- Axes: Les axes sont une partie de la figure où les données sont tracées. Ils incluent les composants tels que les lignes d'axe, les marques de graduation, les étiquettes, etc. Une figure peut contenir plusieurs axes.
- Title (Titre): Le titre est un texte placé en haut des axes pour indiquer de manière succincte le sujet ou la fonction du graphique.
- X-axis label (Étiquette de l'axe X): Texte placé le long de l'axe des x pour identifier le type de données représentées sur cet axe.
- Y-axis label (Étiquette de l'axe Y): Semblable à l'étiquette de l'axe X, mais placé le long de l'axe des y.
- Line (Ligne): Un type de graphique dans lequel les points de données sont reliés par une ligne droite ou courbe. Utilisé souvent pour représenter les tendances sur une série temporelle.
- Markers (Marqueurs): Symboles graphiques (comme des points, des carrés, des cercles) utilisés dans les graphiques de dispersion pour représenter les points de données individuels.
- Legend (Légende): Un élément qui explique les symboles, couleurs ou lignes utilisés dans le graphique pour différencier les groupes de données.
- Grid (Grille): Lignes horizontales et verticales qui aident à mieux évaluer les positions des points de données sur le graphique.
- Major tick label: Les étiquettes associées aux principales graduations sur les axes, généralement plus grandes ou plus fréquentes que les minor ticks.
- Minor tick label: Les étiquettes pour les petites graduations sur les axes, utilisées pour une granularité plus fine.
- Major tick: Les grandes marques sur les axes, généralement à intervalles réguliers pour indiquer des valeurs spécifiques dans l'échelle du graphique.
- Minor tick: Plus petites marques entre les major ticks, offrant une échelle plus détaillée sur le graphique.
- Spine: Les lignes reliant les marques d'axes major ticks et formant la limite du graphique. Les spines peuvent être personnalisées pour changer l'apparence du cadre du graphique.
- X-axis (Axe X) et Y-axis (Axe Y): Les axes horizontaux et verticaux du graphique, respectivement, sur lesquels les données sont tracées. Ils comprennent les lignes d'axe, les ticks, et les étiquettes.

<h1> Création de visualisation avec Maltplotlib à partir d'un réel data set </h1>

```python
import pandas as pd

titanic = pd.read_csv('Titanic-Dataset.csv')
titanic.head(3)
```

<h2> Les courbes </h2>

Objectif : Tracer une ligne montrant quelque chose d'intéressant dans le dataset, par exemple, le prix du billet (fare) en fonction de l'âge

```python
fare_by_age = titanic.groupby(['Age'])['Fare'].mean().reset_index() # reset_index est utile pour recréer un Datframe
fare_by_age
```

```python
fig = plt.figure(figsize=(10,5)) # Définir la taille de la figure
ax = fig.add_subplot(111) # Définir une zone de tracage
ax.plot(fare_by_age['Age'] , fare_by_age['Fare'] , label='Fare by Age')  # Tracer la courbe Fare vs Age
ax.set_title('Prix du ticket en fonction de l\'age')  # Ajouter un titre
ax.set_xlabel('Age')  # Ajouter l'étiquette de l'axe x
ax.set_ylabel('Prix du billet')  # Ajouter l'étiquette de l'axe y
ax.legend()  # Ajouter une légende
plt.show()
```

<b> customisation de la courbe afin de mieux comprendre chaque valeur </b>

- marker : Ce paramètre détermine le style des marqueurs pour chaque point de données sur le graphique (o . , s > )
- color : Ce paramètre définit la couleur des éléments du graphique (- -- -. :)
- linestyle : Ce paramètre spécifie le style de la ligne utilisée pour connecter les points de données sur le graphique
- linewidth : Ce paramètre définit l'épaisseur de la ligne tracée entre les points. 
- markersize : Ce paramètre détermine la taille des marqueurs (les cercles, dans ce cas).

```python
fig = plt.figure(figsize=(10,5)) # Définir la taille de la figure
ax = fig.add_subplot(111) # Définir une zone de tracage
ax.plot(fare_by_age['Age'] , fare_by_age['Fare'] , label='Fare by Age' , color='lightblue', marker='o' , linestyle = '-' , linewidth = 1)  # Tracer la courbe Fare vs Age
ax.set_title('Prix du ticket en fonction de l\'age')  # Ajouter un titre
ax.set_xlabel('Age')  # Ajouter l'étiquette de l'axe x
ax.set_ylabel('Prix du billet')  # Ajouter l'étiquette de l'axe y
ax.grid(True)
ax.legend()  # Ajouter une légende
plt.show()
```

<b> Ventilation de la courbe en fonction de la classe du passager </b>

```python
fare_by_age_class = titanic.groupby(['Age' , 'Pclass'])['Fare'].mean().reset_index()
fare_by_age_class_1 = fare_by_age_class[fare_by_age_class['Pclass'] ==1]
fare_by_age_class_2 = fare_by_age_class[fare_by_age_class['Pclass'] ==2]
fare_by_age_class_3 = fare_by_age_class[fare_by_age_class['Pclass'] ==3]
```

```python
fig = plt.figure(figsize=(10,5)) # Définir la taille de la figure
ax = fig.add_subplot(111) # Définir une zone de tracage
ax.plot(fare_by_age_class_1 ['Age'] , fare_by_age_class_1 ['Fare'] , label='1iere classe' , color='lightblue', marker='o' , linestyle = '-' , linewidth = 1)  
ax.plot(fare_by_age_class_2['Age'] , fare_by_age_class_2['Fare'] , label='2ieme classe ' , color='#33C4FF', marker='o' , linestyle = '-' , linewidth = 1)
ax.plot(fare_by_age_class_3['Age'] , fare_by_age_class_3['Fare'] , label='3ieme classe ' , color='m', marker='o' , linestyle = '-' , linewidth = 1)
ax.set_xlabel('Age')  # Ajouter l'étiquette de l'axe x
ax.set_ylabel('Prix du billet')  # Ajouter l'étiquette de l'axe y
ax.grid(True)
ax.legend()  # Ajouter une légende
plt.show()
```

<h2> Les diagrammes en barres </h2>

<b> Le diagramme en barres simple

Objectif : Tracer le nombre de passagers par classe (Pclass)

```python
passenger_class_counts = titanic['Pclass'].value_counts().sort_index()
passenger_class_counts
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111) 
ax.bar(passenger_class_counts.index , passenger_class_counts.values, color ='blue' ,alpha = 0.8)
ax.set_title('Nombre de Passagers par Classe')
ax.set_xlabel('Classe')
ax.set_ylabel('Nombre de passagers')
ax.set_xticks([1,2,3]) # valeur de chaque diagramme, à faire sinon une échelle numérique va être ajouté au diagramme
plt.show()
```

<h2> Création d'un Graphique Nuage de Points </h2>

<b>  Graphique Nuage de Points Basique </b>

Objectif : Tracer un graphique de nuages de points pour analyser la relation entre l'âge (Age) et le tarif du billet (Fare).

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'] , titanic['Fare'] , alpha = 0.5)
plt.title('Relation entre l’Âge et le Tarif du Billet')
plt.xlabel('Age')
plt.ylabel('Tarif du billet')
plt.show()
```

<b> Graphique Nuage de Points avec catégorie </b>

Objectif : Ajouter des couleurs en fonction de la survie des passagers (Survived) pour mieux visualiser les différences.

```python
colors = titanic['Survived'].map({0:'red' , 1:'green'})
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'] , titanic['Fare'] , alpha = 0.5 , c=colors, label = 'Rouge : Décédé , vert: Survécu')
plt.title('Relation entre l’Âge et le Tarif du Billet')
plt.xlabel('Age')
plt.ylabel('Tarif du billet')
plt.legend()
plt.show()
```

<b> Bubble chart </b>

Objectif : Modifier la taille des points basée sur une troisième variable, comme le nombre de membres de la famille à bord (SibSp + Parch).

```python
family_size = titanic['SibSp'] + titanic['Parch']
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'] , titanic['Fare'] , alpha = 0.5 , c=colors, s = family_size , label = 'Taille de bulle : taille de la famaille' )
plt.title('Âge vs Tarif du Billet par Taille de Famille et Survie')
plt.xlabel('Âge')
plt.ylabel('Tarif du Billet')
plt.legend()
plt.show()
```

<h1> graphique à aires </h1>

<b>  Graphique à Aires Simple </b>

objectif : visualiser les variations du prix des billets (Fare) en fonction de l'âge (Age).

```python
# Préparation des données
fare_by_age = titanic.groupby('Age')['Fare'].mean()

# Création du graphique à aires
plt.figure(figsize=(10, 6))
plt.fill_between(fare_by_age.index , fare_by_age.values , color = "skyblue" ,alpha = 0.4)
plt.title('Prix Moyen des Billets par Âge')
plt.xlabel('Âge')
plt.ylabel('Prix Moyen du Billet')
plt.show()
```

objectif : Créer un graphique représentant les taux de survie des passagers du Titanic, en fonction de leur groupe d'âge et de leur sexe.


1) Calcul du Taux de Survie :
- Calculez le taux de survie moyen pour chaque combinaison de groupe d'âge et de sexe. Utilisez groupby() pour regrouper les données, puis mean() pour calculer la moyenne de la colonne Survived et finalement unstack pour dépivoter le tableau.

2) Création du Graphique :
- Utilisez Matplotlib pour tracer un graphique qui montre les taux de survie pour chaque groupe d'âge, séparés par sexe.
